```python
import re

# Initialisation des tables de données :
DF_ideal_join = DF_ideal.copy()
DF_rvf_join = DF_rvf.copy()

# Définition des fonctions utiles pour le traitement des chaînes de caractères :
def replace_words(s, D_words):
    L_split = s.split(" ")
    L_split = [w if (w not in D_words.keys()) else (D_words[w]) for w in L_split]
    return " ".join(L_split)

def normalize(s):
    s = s.lower()
    s = re.sub(r"[\.\,\!\?\:\;\@\(\)\[\]\+\=\$\€\£\%\~]", "", s)
    s = re.sub(r"[\-\'\_\’]", " ", s)
    s = re.sub(r"[áäàâ]", "a", s)
    s = re.sub(r"[éëèê]", "e", s)
    s = re.sub(r"[íïìî]", "i", s)
    s = re.sub(r"[óöòô]", "o", s)
    s = re.sub(r"[úüùû]", "u", s)
    s = re.sub(r"[ýÿ]", "y", s)
    s = re.sub(r"[ç]", "c", s)
    s = re.sub(r"[ñ]", "n", s)
    s = re.sub(r"[œ]", "oe", s)
    s = s.replace('&', ' et ')
    s = re.sub(r"[\s]+", " ", s)
    s = s.strip()
    s = replace_words(s, {
        'gewurtraminer': 'gewurztraminer', 
        'ermitage': 'hermitage',
        'vt': 'vendanges tardives',
        'sgn': 'selection de grains nobles',
        'vv': 'vieilles vignes',
        '1er': 'premier',
        'vdp': 'vin de pays',
        'vdt': 'vin de table',
        'st': 'saint',
        'ste': 'sainte'
    })
    return s

def remove_articles(s):
    L_articles = ['le', 'la', 'les', 'de', 'des', 'du', 'l', 'd']
    return " ".join([word for word in s.split(' ') if word not in L_articles])

def remove_sub(s, L_sub):
    for w in L_sub:
        s = s.replace(" "+w+" ", " ")
        s = re.sub(r"\s"+w+r"$", "", s)
        s = re.sub(r"^"+w+r"\s", "", s)
        s = re.sub(r"^"+w+r"$", "", s)
    return s

def change_app(s):
    # Gestion des appellations qui ont changé de nom ou qui se chevauchent :
    s = s.replace("listrac medoc", "listrac")
    s = s.replace("castillon cotes de bordeaux cotes de castillon", "cotes de castillon")
    s = s.replace("castillon cotes de bordeaux", "cotes de castillon")
    s = s.replace("barsac", "sauternes")
    s = s.replace("mazy chambertin", "mazis chambertin") 
    s = s.replace("coteaux du languedoc pic saint loup", "coteaux du languedoc") 
    s = s.replace("pic saint loup", "coteaux du languedoc")
    s = s.replace("coteaux du languedoc la clape", "coteaux du languedoc")  
    s = s.replace("montlouis sur loire", "montlouis")
    s = s.replace("vin de pays bouches du rhone", "vin de pays des bouches du rhone") 
    s = s.replace("igp saint guilhem le desert cite d aniane", "igp pays d herault")
    s = s.replace("igp saint guilhem le desert", "igp pays d herault")
    s = s.replace("vin de pays de l herault", "igp pays d herault")
    
    return s  

def remove_useless_app(s):
    L_useless = ["grand cru", "premier cru", "vendanges tardives", "selection de grains nobles", "blanc", "rouge", "rose", "classe"]
    s = remove_sub(s, L_useless)
    s = change_app(s)
    return s

def remove_useless_dom(s):
    L_useless = ['domaine du', 'domaine des', 'domaine de', 'domaines', 'domaine', 'chateau de', 'chateau de', 'chateau des', 'chateau', 'mas de', 'mas', 'maison',
                 'longueville', 'selections parcellaires', 'heritiers', 'sauternes', 'comte de saint victor', 'famille portalis', 'a pomerol', 'rausan segla',
                 'artemis', 'rene engel', 'frederic pourtalie', 'jerome bressy', 'laurent vaille', 'eloi durrbach', 'charles hours', 'eric et charles eric boisseaux',
                 'christine et eric nicolas', 'paul et laurence feraud', 'bruno et claudie bilancini', 'alain dominique perrin', 'pascal et jean marc verhaeghe',
                 'jean pierre et francois perrin', 'herve bizeul', 'famille tardieu', 'famille tari', 'famille peyraud',  
                 'famille richard', 'famille sabon', 'famille gonnet', 'famille brunel', 'andre brunel', 'sophie et catherine armenier',
                 'famille abeille fabre', 'jean paul daumen', 'vignobles brunier', 'mommessin', 'marlene soria', 'philippe foreau',
                 'chantal lecouty et jean claude le brun', 'famille guibert de la vaissiere', 'chavigne', 'famille rougier',
                 'paul coulon et fils', 'jean et jean paul versino', 'aime sabon', 'sylvie vacheron', 'christophe peyrus et françoise julien'
                 'millesime', 'vintage', 'brut', 'blanc de blancs', 'et fils', 'cuvee', 'grand cru', 'cave', 'liquoreux', 'lieu dit']
    s = remove_sub(s, L_useless)
    s = change_app(s)

    # Quelques domaines connus qui ont changé de nom ou qui ont plusieurs noms :
    s = s.replace("alesme becker", "alesme")
    s = s.replace("saint pierre saint julien", "saint pierre")
    s = s.replace("saint pierre medoc", "saint pierre")
    s = s.replace("saint pierre pomerol", "saint pierre")
    s = s.replace("caillou barsac", "caillou")
    s = s.replace("naudin ferrand claire naudin", "naudin ferrand")
    s = s.replace("henri naudin ferrand", "naudin ferrand")
    s = s.replace("guadet saint julien", "guadet")
    s = s.replace("lagrange pomerol", "lagrange")
    s = s.replace("jack confuron cotetidot", "confuron cotetidot")
    s = s.replace("pierre et jerome coursodon", "coursodon") 
    s = s.replace("didier dagueneau", "dagueneau") 
    s = s.replace("moet et chandon dom perignon", "dom perignon") 
    s = s.replace("remizieres cave desmeure", "remizieres") 
    s = s.replace("jean michel gerin", "gerin") 
    s = s.replace("alain graillot", "graillot") 
    s = s.replace("laurent pere", "dominique laurent") 
    s = s.replace("bernard dugat py", "dugat py") 
    s = s.replace("henri bonneau", "henri bonneau") 
    s = s.replace("famille hugel", "hugel") 
    s = s.replace("jean trapet pere", "trapet") 
    s = s.replace("jean trapet", "trapet") 
    s = s.replace("la coulee de serrant", "coulee de serrant") 
    s = s.replace("michel lafarge", "lafarge") 
    s = s.replace("leroy sa", "leroy") 
    s = s.replace("louis carillon", "jacques carillon")     
    s = s.replace("lynsolence vignobles denis barraud", "lynsolence")      
    s = s.replace("montus la tyre", "vignobles alain brumont")     
    s = s.replace("montus prestige", "vignobles alain brumont")     
    s = s.replace("montus", "vignobles alain brumont")     
    s = s.replace("bouscasse", "vignobles alain brumont")     
    s = s.replace("centre vinicole nicolas feuillatte", "nicolas feuillatte")     
    s = s.replace("nicolas feuillate", "nicolas feuillatte") 
    s = s.replace("clos saint landelin veronique et thomas mure", "r mure")
    s = s.replace("clos saint landelin", "r mure")       
    s = s.replace("salon cuvee s", "")     
    s = s.replace("pierre usseglio", "usseglio")     
    s = s.replace("jean pierre et thierry usseglio", "usseglio")     
    s = s.replace("la vieille julienne", "vieille julienne")
    s = s.replace("rene et vincent dauvissat", "vincent dauvissat")
    s = s.replace("anne et jean francois ganevat", "ganevat")
    s = s.replace("jean francois ganevat", "ganevat")
    s = s.replace("christine schlumberger", "schlumberger")
    s = s.replace("armailhac mouton baronne philippe", "armailhac")
    s = s.replace("rayas reynaud", "rayas")
    s = s.replace("michel chapoutier", "chapoutier")
    s = s.replace(" m chapoutier", " chapoutier")
    s = re.sub(r"^m chapoutier", "chapoutier", s)
    s = s.replace("remizieres desmeure", "remizieres")
    s = s.replace(" e guigal", " guigal")
    s = re.sub(r"^e guigal", "guigal", s)
    s = s.replace("laurent ponsot", "ponsot")
    s = s.replace("meo camuzet frere et soeurs", "meo camuzet")
    s = s.replace("olivier jullien", "jullien")
    s = s.replace("jean et jean louis trapet", "trapet")
    s = s.replace("la grange des peres", "grange des peres")
    s = s.replace("rose grand siecle alexandra", "alexandra")
    s = s.replace("alexandra grande rose", "alexandra")
    s = s.replace("sir winston churchill", "winston churchill")
    s = s.replace("christophe roumier", "georges roumier")
    s = s.replace("belair belair monange", "belair monange")
    s = s.replace("le pin", "pin")
    s = s.replace("les ormes de pez", "ormes de pez")
    return s

def remove_double(s):
    L_words = s.split(" ")
    L_words = [L_words[0]] + [L_words[i] for i in range(1, len(L_words)) if L_words[i]!=L_words[i-1]]
    return " ".join(L_words)

# Constitution d'une clé de jointure correspondant au nom de domaine :
DF_ideal_join['join_domain'] = DF_ideal_join['domaine'].apply(normalize).apply(remove_useless_dom)
DF_rvf_join['join_domain'] = DF_rvf_join['domain_name'].apply(normalize).apply(remove_useless_dom)

# Constitution d'une clé de jointure correspondant à l'appellation :
DF_ideal_join['join_appellation'] = DF_ideal_join['appellation'].fillna('').apply(normalize).apply(remove_useless_app)
DF_ideal_join['join_appellation'][DF_ideal_join.pays_region=='Alsace'] = "alsace"
DF_rvf_join['join_appellation'] = DF_rvf_join['appellation'].fillna('').apply(normalize).apply(remove_useless_app)
DF_rvf_join['join_appellation'][DF_rvf_join.region=='Alsace'] = "alsace"

# Constitution d'une clé de jointure correspondant au nom du vin :
DF_ideal_join['join_wine'] = DF_ideal_join.apply(
    lambda S: remove_articles(remove_double(remove_sub(
        remove_useless_dom(normalize(S['nom_du_vin'])),
        [S['join_appellation'], S['join_domain']]
    ))),
    axis=1
)
DF_rvf_join['join_wine'] = DF_rvf_join.apply(
    lambda S: remove_articles(remove_double(remove_sub(
        remove_useless_dom(normalize(S['wine_name'])),
        [S['join_appellation'], S['join_domain']]
    ))),
    axis=1
)

# Constitution d'une clé de jointure correspondant à la couleur (rouge, rose ou blanc) :
DF_ideal_join['join_color'] = DF_ideal_join['couleur'].apply(lambda s: 
    s.split()[0].lower().replace('é', 'e') if s.startswith('R') else 'blanc'
)
DF_rvf_join['join_color'] = DF_rvf_join['vintage_twitterimg'].fillna('').apply(lambda s:
    re.sub(r"[^\-]*\-", "", re.sub(r"\..*$", "", s)) 
)

# Mise au propre des millésimes :
DF_ideal_join['join_vintage'] = DF_ideal_join['millesime']
DF_rvf_join['join_vintage'] = DF_rvf_join['vintage'].fillna(0).astype(int)

# Réinitialisation des indexes :
DF_ideal_join = DF_ideal_join.reset_index(drop=True).reset_index()
DF_rvf_join = DF_rvf_join.reset_index(drop=True)
```

# Exécution et test de la jointure

```python
###################################################
# Exécution de la jointure :

# Pour trouver la bonne jointure, on ne garde qu'un exemplaire de chaque vin millésimé côté RVF, même s'il a plusieurs notes :
import re
print(pd.__version__)  # 1.1 au moins nécessaire pour le groupby ci-après
L_wine_col = ['domain_name', 'domain_stars', 'wine_name', 'region', 'appellation', 'vintage_twitterimg', 'wine_types', \
              'wine_varieties', 'vintage', 'vintage_types', 'vintage_varieties', 'vintage_prices']
DF_rvf_join2 = DF_rvf_join.groupby(L_wine_col, dropna=False, as_index=False).first()
#display(DF_rvf_join2.head())

L_join_col = ['join_domain', 'join_appellation', 'join_wine', 'join_color', 'join_vintage']
DF_merge0 = DF_ideal_join[['index']+L_join_col].merge(DF_rvf_join2[L_join_col], how='left', on=['join_appellation', 'join_color', 'join_vintage', 'join_domain'], suffixes=('_ideal', '_rvf'))
print(DF_merge0.columns.tolist())

# Fonction définissant un score de correspondance des noms du vin, entre 0 (KO) et 1 (OK) :
def wine_match(w1, w2):
    if type(w2).__name__!='str':
        return 0
    elif w1==w2:
        return 1
    elif w2=='':
        return 0
    elif w1=='':
        return 0        
    else:
        L_w1 = w1.split(" ")
        L_w2 = w2.split(" ")
        score1 = sum([int(word in L_w1) for word in L_w2]) / len(L_w2)
        score2 = sum([int(word in L_w2) for word in L_w1]) / len(L_w1)
        return 0.9*score1*(0.8+0.2*score2)  # Le score 1 semble beaucoup plus significatif que le score 2, d'où cette pondération

print("Nombre de vins côté Idealwine :", len(DF_ideal_join))
print("Nombre d'associations Idealwine-RVF possibles à ce stade :", len(DF_merge0))
DF_merge0['match_score'] = DF_merge0.apply(lambda S: wine_match(S['join_wine_ideal'], S['join_wine_rvf']), axis=1)
```

```python
# Examen des vins sans correspondance :
# Rien de choquant : la RVF, malheureusement, ne goûte pas toutes les cuvées quand un vigneron en a beaucoup...

L_ideal = ['index', 'join_domain', 'join_appellation', 'join_wine_ideal', 'join_color', 'join_vintage']
DF_failure = DF_merge0.groupby(L_ideal, as_index=False)['match_score'].max()
print(len(DF_failure))
DF_failure = DF_failure[DF_failure.match_score==0]
DF_failure = DF_failure[DF_failure.join_vintage>=2000]  # Les millésimes antérieurs à 2000 sont moins présents côté RVF, normal
print(len(DF_failure))

#display(DF_failure[DF_failure.join_appellation=='pauillac'].head(30))
#display(DF_rvf[DF_rvf.domain_name.str.startswith("Château Pichon")])
display(DF_failure.sample(30, random_state=42))
```

```python
# Traitement des correspondances multiples :
score_threshold = 0.75
DF_merge1 = DF_merge0[DF_merge0.match_score > score_threshold]
print(len(DF_merge1))

L_ideal = ['index', 'join_domain', 'join_appellation', 'join_wine_ideal', 'join_color', 'join_vintage']
#DF_test = DF_merge1.groupby(L_ideal, as_index=False)[['index_rvf']].count()[['index_ideal', 'index_rvf']]
#L_doublons = DF_test[DF_test.index_rvf>1]['index_ideal'].tolist()
#print(len(L_doublons))

DF_merge2 = DF_merge1.sort_values(by='match_score', ascending=False)
DF_merge2 = DF_merge2.groupby(L_ideal, as_index=False).first().sort_values(by='match_score')
print("Nombre de correspondances trouvées :", len(DF_merge2))
DF_merge2.head(80).sample(20)
# Le seuil autour de 0.75 semble pertinent, avec inévitablement un peu de "faux positifs" et un peu de "faux négatifs"
```

```python
# On peut donc constituer la table de correspondance entre un n° d'index côté Idealwine et les clés de jointures acceptées côté RVF :
DF_join_table = DF_merge2[['index', 'join_wine_rvf']].rename({'join_wine_rvf': 'join_wine'}, axis=1)
DF_ideal_wine_rvf = DF_ideal_join.drop(columns=['join_wine']).merge(DF_join_table, how='left', on=['index'])

DF_rvf_join_bis = DF_rvf_join.copy()
DF_rvf_join_bis.columns = pd.Series(DF_rvf_join.columns).apply(lambda s: s if (s.startswith('join_') or s.startswith('rvf_')) else 'rvf_'+s)
# print(DF_rvf_join.columns.tolist())

L_join = ['join_domain', 'join_appellation', 'join_color', 'join_vintage', 'join_wine']
DF_joined = DF_ideal_wine_rvf.merge(DF_rvf_join_bis, how='left', on=L_join)
DF_joined = DF_joined.drop(columns=L_join+['index'])
print("Nombre de lignes de la base après jointure :", len(DF_joined))
# Remarque : une ligne de la base Idealwine est dupliquée autant de fois qu'on trouve de notes pour l'unique vin correspondant côté RVF

#DF_joined.head(20)
print(DF_joined.columns.tolist())
```
