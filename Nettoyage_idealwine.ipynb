# Chargement et examen des données

```python
import pandas as pd
import glob
import re

path = r'.\scrapy\idealwine\data\*.json'

L_files = list()
for filename in glob.glob(path):
    DF_part = pd.read_json(filename)
    L_files.append(DF_part)

DF_ideal0 = pd.concat(L_files, axis=0, ignore_index=True).sample(frac=1)
DF_ideal0 = DF_ideal0.drop_duplicates(subset=DF_ideal0.columns[1:7], keep='last')
```

# Nettoyage

```python
DF_ideal = DF_ideal0.copy()

# On ne garde que les vins d'un domaine identifié, car les autres n'existent pas vraiment :
DF_ideal = DF_ideal[~DF_ideal.domaine.isna()]

# On ne garde que les vins français et même les corses :
""" display(DF_ideal.pays_region.value_counts().index) """
L_keep = ['Bordeaux', 'Bourgogne', 'Vallée du Rhône', 'Vallée de la Loire', 'Champagne', 'Alsace', 'Languedoc',
          'Provence', 'Sud-Ouest', 'Jura', 'Roussillon', 'Beaujolais', 'Corse', 'Savoie', 'Charente']
DF_ideal = DF_ideal[DF_ideal.pays_region.apply(lambda reg : (reg in L_keep))]

# On supprime les alcools et assortiments qui ne sont pas des vins :
DF_ideal = DF_ideal[DF_ideal.appellation.apply(lambda s : (re.search(r"ognac", s)==None))]
DF_ideal = DF_ideal[DF_ideal.appellation.apply(lambda s : (re.search(r"rmagnac", s)==None))]
DF_ideal = DF_ideal[DF_ideal.appellation.apply(lambda s : (re.search(r"artreuse", s)==None))]
DF_ideal = DF_ideal[DF_ideal.appellation.apply(lambda s : (re.search(r"acvin", s)==None))]
DF_ideal = DF_ideal[DF_ideal.appellation.apply(lambda s : (re.search(r"hisky", s)==None))]
DF_ideal = DF_ideal[DF_ideal.couleur!='Divers']

# On supprime les vins sans millésime, dont l'âge et la capacité de conservation ne peuvent pas être maîtrisés :
DF_ideal = DF_ideal[(DF_ideal.millesime!='----')]

# On supprime la première colonne inutile (donnée de scraping) et on range les colonnes suivantes :
L_col_right = DF_ideal.columns[7:].tolist()
L_col_right.sort()
L_col = DF_ideal.columns.tolist()[1:7] + L_col_right
DF_ideal = DF_ideal[L_col]

# On supprime les notes (évaluations) non exploitables : celle d'Idealwine (modalités de constitution inconnue)
# et celles rarement renseignées :
DF_ideal = DF_ideal.drop(columns=['note_iDw', 'note_DECANTER', 'note_Lepoint', 'note_RVF'])

print(DF_ideal.info())
```

```python
# Sauvegarde :
import time
DF_ideal.to_csv("clean_data/idealwine_" + str(int(time.time())) + ".csv", index=False)


```
